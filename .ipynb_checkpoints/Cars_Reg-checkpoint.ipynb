# EDA

```python
#View unique values for a categorical variables
for col in df.select_dtypes('object').drop(['Invoice','MSRP'],axis=1):
    print(col, ':',df[col].nunique())
```

```python
def string_currency(df,col):
    df[col] = df[col].apply(lambda x: ''.join(x[1:].split(','))).astype('int64')
```

```python
#Convert Invoice and MSRP column to numeric and remove Currency '$'
string_currency(df,'Invoice')
string_currency(df,'MSRP')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
for cols in df.select_dtypes(include='number'):
    plt.figure()
    sns.distplot(df[cols].dropna())
```

```python
colours = ['red','green','blue','cyan','yellow','magenta']
df.groupby('Make')['MSRP'].mean().sort_values(ascending=False).head(10).plot(kind='barh',color = colours)
plt.title('Top 10 Make Average Price')
plt.xlabel('Price')
```

```python
#Calculate Mark-Up on each vehicle
df['Mark_up'] = df['MSRP'] - df['Invoice']
df['Mark_up_%'] = df['Mark_up'] / df['Invoice'] * 100
df.head()
```

```python
#Create average line
mark_up_avg = [np.mean(df['Mark_up_%'])]*15
mark_up_avg

#Calculate which vehicle Make has the highest Mark-up
round(df.groupby('Make').mean()['Mark_up_%'].sort_values(ascending=False),2).head(15).plot(kind='bar',color='orange',edgecolor='black',fontsize=12)
#plt.plot(df['Mark_up_%'],mark_up_avg, linestyle='--',label='Average',color='black')
plt.plot(mark_up_avg,linestyle='--',label='Average',color='black')
plt.title('Top 15 Highest Mark-Ups')
plt.xlabel('Price',)
plt.xticks(rotation=60)
plt.ylabel('Avg_Mark-Up')
```

```python
#Compare Average Mark-up to average price of top 15 Vehicles
makes_15 =[]
avg_price =[]
list_to_markup = df.groupby('Make').mean()['Mark_up_%'].sort_values(ascending=False).head(15).index
for mk in list_to_markup:
    makes_15.append(mk)
    avg_price.append(round(df[df['Make']==mk]['MSRP'].mean(),2))
    #print(mk,':',round(df[df['Make']==mk]['MSRP'].mean()),2)
df_15= pd.DataFrame(zip(makes_15,avg_price),columns=['Makes','Avg_price'])
#df_15.plot(x=df_15['Makes'],y=df_15['Avg_price'],kind='bar')
sns.barplot(x=df_15['Makes'],y=df_15['Avg_price'])
plt.xticks(rotation=60)
```

```python
#Correlation between "MSRP" & "Mark_up%"
sns.heatmap(df[['MSRP','Mark_up_%']].corr(),annot=True)
plt.ylim(0,2)
```

- On visualising top 15 marked up vehicles
- We could draw conclusion that The higher the price of the vehicle, the higher the mark-up
- On visualizing the the average price of the top 15 and the correlation between the variables, we can see the there is no positive relationship.

```python
#Average Mark-up percentage per Vehicle type
df.groupby('Type').mean()['Mark_up_%'].plot(kind='bar',color=['red','blue','yellow','green','cyan'])
plt.xticks(rotation=60)
plt.title('Percenage Mark-up per Vehicle Type')
plt.ylabel('Mark_up Percentage_%')
```

```python
#Create DataFrame with the average per Origin
df_orig = pd.DataFrame(df.groupby('Origin').mean()['Invoice']).reset_index()
df_orig
#Viewing the origin of Vehicles within Dataset.

#add colors
colors = ['#ff9999','#66b3ff','#99ff99']
# only "explode" the 2nd slice (i.e. 'Hogs')
explode = (0.1, 0,0)  
fig1, ax1 = plt.subplots(figsize=(7,6))
ax1.pie(df_orig['Invoice'],explode=explode,autopct='%1.1f%%',colors=colors,
        shadow=True, startangle=90,textprops={'fontsize': 14})
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.title('Origin of Vehicles',fontsize=20,fontdict={'family': 'serif'})
plt.legend(df_orig['Origin'],loc=3,fontsize =8)
plt.show()
```

```python
#Compare the average price for DriveTrain
df.groupby('DriveTrain').mean()['Invoice'].plot(kind='bar')
plt.title('Avg Price per DriveTrain',fontsize=20,fontdict={'family': 'serif'})
plt.ylabel('Price')
plt.xticks(rotation = 60)
plt.show()
```

- As per DriveTrain bar chart, we can see that Front Drive Vehicles tend to be cheaper and rear wheel drive to be more priced.

```python
#Create DataFrames with Horsepower and Invoice
df_horse = pd.DataFrame(df.groupby('DriveTrain').mean()['Horsepower']).reset_index()
df_price = pd.DataFrame(df.groupby('DriveTrain').mean()['MSRP']).reset_index()#.plot(kind='bar')

fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(111)
ax2 = ax.twinx()

width = 0.4

df_horse.Horsepower.plot(kind='bar',color='red',ax=ax,width=width,position=1,label='Horsepower')
df_price.MSRP.plot(kind='bar',color='blue',ax=ax2,width=width,position=0,label='MSRP')

ticks= [0,1,2]
drive=['All', 'Front', 'Rear']
ax.set_ylabel('HorsePower')
ax.legend(loc=2)
ax2.set_ylabel('Price')

ax2.set_ylim(top=50000,bottom=0)
ax2.set_xticks(ticks)
ax2.set_xticklabels(drive,rotation=60)
ax2.legend(loc='best')
plt.show()
```

- Grouping the DriveTrain, as per the graph above, the price has a positive relationship with Horsepower

# Preprocessing

```python
#Missing Values
df.isnull().sum()
fill_cyl = df[(df['Make']=='Mazda') & (df['Type']=='Sports')]['Cylinders'].mode()[0]
df['Cylinders'] = df['Cylinders'].fillna(fill_cyl)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
#View Outliers
for col in df.columns:
    plt.figure(figsize=(6,4))
    if df[col].dtype!= 'object':
        sns.boxplot(df[col])
```

```python
df_dummy = pd.get_dummies(df,columns=['Make','Model','Type','Origin','DriveTrain'],drop_first=True)
```

```python
df_dummy.head()
```

```python
df_dummy = df_dummy.drop(['Invoice','Mark_up','Mark_up_%'],axis=1)
```

```python
df_dummy.head()
```

```python
df_dummy.shape
```

```python
X = np.array(X)
y = np.array(y)
```
